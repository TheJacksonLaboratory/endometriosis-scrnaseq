```python
import pandas as pd
import numpy as np 

import sys
sys.path.append("/projects/robson-lab/research/endometriosis/manuscript/code/plotting")
sys.path.append("/projects/robson-lab/research/endometriosis/manuscript/code/external")
from volcano import volcano_plot
```

```python
main_dir = "/projects/robson-lab/research/endometriosis/"
sample_id = "Endo-Tissue-EC19001-SC2100428-3"
```

```python
Eut_deg = pd.read_csv(f"{main_dir}analysis/{sample_id}/bulkRNAseq/DEG-Eutopic-SCvsBS_20220106.csv",
                         index_col  = "genes")

Ect_deg = pd.read_csv(f"{main_dir}analysis/{sample_id}/bulkRNAseq/DEG-Ectopic-SCvsBS_20220106.csv",
                         index_col  = "genes")

Ovr_deg = pd.read_csv(f"{main_dir}analysis/{sample_id}/bulkRNAseq/DEG-Ovary-SCvsBS_20220106.csv",
                         index_col  = "genes")
```

```python
deg_lists = [Eut_deg, Ect_deg, Ovr_deg]
names = ["Eutopic", "Peritoneal endometriosis", "Ovarian endometrosis"]
genes_highlight = [["HBA2","HBB","HBA1","PTPRS","SHANK1"], #eutopic
                   ["PTPRS","SYNPO","TNS1","PLIN1","PLIN4","HBB","LPL","FLNA"], #ectopic
                   ["FLNA","PTPRS","HBB","SYNPO"]] #ovary
for lists, name, genes in zip(deg_lists, names, genes_highlight):
    volcano_plot(lists, "",lfc=3,
                 neg_label="single-cell", pos_label="bulk-tissue", 
                 label_genes = True,
                 genes_to_label= genes,
                 top_neg_genes= 10, top_pos_genes= 10,
                 filename=f"{main_dir}figures/manuscripts/v4/Fig_S1/Volcano-{name}-SCvsBS.pdf"
                )
```
